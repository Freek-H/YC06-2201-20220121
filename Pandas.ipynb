```python
data_nieuwe_namen['UNIT'].unique()
```

```python
data_nieuwe_namen['UNIT'][6]
```

```python
nieuwe_waarden = []
for index, row in data_nieuwe_namen.iterrows():
    if row['UNIT'].startswith('Microgram'):
        nieuwe_waarden.append( row['mg/100g'] / 1000 )
    else:
        nieuwe_waarden.append(row['mg/100g'])
nieuwe_waarden
```

```python
data_aangepast = data_nieuwe_namen.copy()
data_aangepast['mg/100g'] = nieuwe_waarden
data_aangepast.head()
```

```python
data_nieuwe_namen.describe()
```

```python
data_aangepast.describe()
```

```python
data_aangepast.head()
```

```python
data_dropped = data_aangepast.drop(columns=['FOOD_ID', 'code', 'COUNTRY', 'efsaprodcode2',
       'level1', 'level2', 'level3', 'NUTRIENT_ID', 'UNIT'])
data_dropped.head()
```

```python
data_dropped.set_index('Voedsel naam')
```

```python
data_pivoted = data_dropped.pivot_table(index='Voedsel naam', columns='Voedingsstof', values='mg/100g')
data_pivoted.head()
```
